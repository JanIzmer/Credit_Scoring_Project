Imports

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os
sys.path.append(os.path.abspath('..'))
from src.preprocessing import clean_credit_data
```

Read CSV and clear data

```python
df = pd.read_csv('../data/raw/cs-training.csv', index_col=0)
df_clean = clean_credit_data(df)
```

NAN check all must be 0

```python
df_clean.isna().sum()
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

# 1   
plt.figure(figsize=(10,5))
sns.histplot(df_clean['MonthlyIncome'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.show()

# 2 Boxplot  
plt.figure(figsize=(10,3))
sns.boxplot(x=df_clean['MonthlyIncome'], color='lightgreen')
plt.title('Monthly Income Boxplot')
plt.xlabel('Monthly Income')
plt.show()

# 3 Boxplot  
plt.figure(figsize=(10,5))
sns.boxplot(x='SeriousDlqin2yrs', y='MonthlyIncome', data=df_clean)
plt.title('Income vs Default Status')
plt.xlabel('Default (0=No, 1=Yes)')
plt.ylabel('Monthly Income')
plt.show()

# 4 Violin plot
plt.figure(figsize=(10,5))
sns.violinplot(x='SeriousDlqin2yrs', y='MonthlyIncome', data=df_clean)
plt.title('Income Distribution by Default Status')
plt.xlabel('Default (0=No, 1=Yes)')
plt.ylabel('Monthly Income')
plt.show()

# 5 Log- 
df_clean['MonthlyIncome_log'] = df_clean['MonthlyIncome'].apply(lambda x: np.log1p(x))
plt.figure(figsize=(10,5))
sns.histplot(df_clean['MonthlyIncome_log'], bins=50, kde=True, color='salmon')
plt.title('Log-Transformed Monthly Income')
plt.xlabel('Log(Monthly Income + 1)')
plt.ylabel('Count')
plt.show()
```
